# 연습문제 - 당뇨병 다중회귀

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

- 당뇨병 데이터 셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

- 산점도와 선형회귀 직선

```python
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(16,8))
for i, feature in enumerate(diabetes.feature_names):
    row, col = i//5, i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- target vs 각 feature

```python
from sklearn.linear_model import LinearRegression
```

```python
y = diabetes.target
for feature in diabetes.feature_names:
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X, y)
    r2 = lr.score(X, y)
    print(f'{feature}:\t{r2:.4f}')
```

- train/test dataset 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data, diabetes.target, test_size=0.1, random_state=2021
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

- Linear Regression 모델

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)
```

```python
# Weight
lr.coef_
```

```python
# bias
lr.intercept_
```

- Decision Tree Regressor 모델

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
pred_dt = dtr.predict(X_test)
```

- 결과 비교

```python
from sklearn.metrics import mean_squared_error
mse_lr = mean_squared_error(y_test, pred_lr)
mse_dt = mean_squared_error(y_test, pred_dt)
print(f'MSE - LR: {mse_lr:.4f}, DTR: {mse_dt:.4f}')
```

```python
res = pd.DataFrame({
    'y_test':y_test, 'lr':pred_lr, 'dt':pred_dt
})
res.head(10)
```
